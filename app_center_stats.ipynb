```python
import requests
import json
response_API = requests.get('https://status.appcenter.ms/api/v2/incidents.json')
```

```python
print(response_API)
```

```python
data = json.loads(response_API.text)
```

```python
data['incidents'][0]
```

```python
newArray= []
for i in data['incidents']:
  temparr = {'id':i['id'],'status':i['status'], 'name':i['name'], 'impact':i['impact'],'Created_at':i['created_at']}
  if(i['incident_updates'][0]['affected_components'] != None):
    temparr['affected_components'] = [j['name'] for j in i['incident_updates'][0]['affected_components']]
  else:
    temparr['affected_components'] = 'None'
  newArray.append(temparr)
  # newArray.append({'id':i['id'],'status':i['status'], 'name':i['name'], 'impact':i['impact'],'affected_components':i['incident_updates'][0]['affected_components'] })

```

```python
import pandas as pd
df = pd.DataFrame(newArray)
df.head()
```
