```
import numpy as np
from PIL import Image

im = np.array(Image.open('image.jpg').convert('L')) #you can pass multiple arguments in single line
print(im.shape)
```

```
l=[]
for i in im:
  for j in i:
    if(j>0):
      l.append(1)
    else:
      l.append(0)
```

```
l
```

```
s=[]
count1=0
count0=0
for i in range(len(l)):
  if(l[i]==1):
    count1=count1+1
    if(count1==3):
      count0=0
  else:
    count0=count0+1
    if(count0==3):
      count1=0
  if(count1==13):
    s.append(1)
    count1=0
    count0=0
  if(count0==13):
    s.append(0)
    count0=0
    count1=0
```

```
print(s)
```

```
n=''.join(map(str, s))
f = open("demofile2.txt", "a")
f.write(str(n))
f.close()
```
